```python
import re
def clean_markdown(text):
    text = re.sub(r'\\\*\\\*', '', text)
    text = re.sub(r'\\\*', '', text)
    return text
```

```python
new = []
for x in df["final"]:
    new.append(clean_markdown(x))
df["cleaned"] = new
```

```python
import os

for i, x in enumerate(df["cleaned"]):
    with open(r"C:\Users\91962\voluntas2\rfps_txt\train\\" + f"rfp_{i+1}.txt", "w", encoding="utf-8") as f:
        f.write(x)
```
