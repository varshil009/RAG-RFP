### copy common files from all parts to new folder

```python
import shutil
```

```python
i = 0
for f in k:
    for part in parts:
        
        source_folder_path = os.path.join(folder, part, f)
        destination_folder_path = os.path.join(new_path, f)
        if not os.path.exists(source_folder_path):
            print(f"Skipping missing folder: {source_folder_path}")
            continue
            
        pdfs = os.listdir(source_folder_path)
        for file in pdfs:
            source_path = os.path.join(source_folder_path, file)
            dest_path = os.path.join(destination_folder_path, file)
            if os.path.isfile(source_path):
                shutil.copy2(source_path, dest_path)
                print(i, f"Moved: {file}")
                i +=1
```
